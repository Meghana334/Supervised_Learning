### Perceptron Learning Concept:
Train weights over epochs and iterate prediction error back to the perceptron for each epoch.

During the error propagation process, we will make changes in our weight values till the time our perceptron makes accurate predictions.

Weight(new) = Weight(old) - Learning rate (Alpha) * (Gradient of error with respect to weight)

Error = output - predicted_output
